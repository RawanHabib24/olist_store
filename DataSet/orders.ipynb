```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
```

```python
orders.fillna(pd.to_datetime('1999-01-01 00:00:00') , inplace=True)
```

```python
orders['purchase_approved_diff'] = (orders['order_approved_at'] - orders['order_purchase_timestamp'])
```

```python
orders['purchase_approved_diff'] = pd.to_timedelta(orders['purchase_approved_diff'])
orders['hours_to_approve'] = (orders['purchase_approved_diff'].dt.components['hours'] + orders['purchase_approved_diff'].dt.components['days']*24)
orders['minutes_to_approve'] = orders['purchase_approved_diff'].dt.components['minutes']
orders['sec_to_approve'] = orders['purchase_approved_diff'].dt.components['seconds']
```

```python
orders
```

```python
orders_without_nulls = orders[orders['order_approved_at'] != pd.to_datetime('1999-01-01 00:00:00')]
avg_approved_time = orders_without_nulls['purchase_approved_diff'].mean()
```

```python
avg_approved_time
```

```python
delivered_orders.isna().sum()
```

```python
delivered_approved_nulls = (orders['order_status'] == 'delivered') &(orders['order_approved_at']==pd.to_datetime('1999-01-01 00:00:00'))
```

```python
# fill approving date nulls when order_status is approved with the average approving time
orders.loc[delivered_approved_nulls, 'order_purchase_timestamp'] + avg_approved_time
```

```python
orders.loc[delivered_approved_nulls, 'order_approved_at']
```

fill carrier date nulls with delivered orders using avg differernce between approving and carrier without nulls

```python
orders
```

```python
avg_carrier_time_difference_without_nulls = orders[(orders['order_delivered_carrier_date'] != pd.to_datetime('1999-01-01 00:00:00')) 
                                                   & (orders['order_approved_at'] != pd.to_datetime('1999-01-01 00:00:00'))]
```

```python
carrier_difference = 
avg_carrier_time_difference_without_nulls['order_delivered_carrier_date'] - avg_carrier_time_difference_without_nulls['order_approved_at']
carrier_difference
```

```python
avg_carrier_time =carrier_difference.mean()
avg_carrier_time
```

```python
# fill carrier nulls when order_status is delivered with the average carrier tim
delivered_carrier_nulls = (orders['order_status'] == 'delivered') & (orders['order_delivered_carrier_date'] == pd.to_datetime('1999-01-01 00:00:00'))
orders.loc[delivered_carrier_nulls, 'order_delivered_carrier_date'] = orders.loc[delivered_carrier_nulls, 'order_approved_at']+avg_carrier_time
```

```python
customers_orders =pd.merge(customers,orders,how='left', on = 'customer_id')
customers_orders
```

```python
customers_orders.isna().sum()
```

```python
delivered_customers_no_nulls = (orders['order_status'] == 'delivered') & (orders['order_delivered_customer_date'] != pd.to_datetime('1999-01-01 00:00:00')) & (orders['order_delivered_carrier_date'] != pd.to_datetime('1999-01-01 00:00:00'))
customers_carrier_difference = orders[delivered_customers_no_nulls]['order_delivered_customer_date'] - orders[delivered_customers_no_nulls]['order_delivered_carrier_date']
customers_carrier_difference.isna().sum()
```

```python
orders['customers_carrier_difference'] = customers_carrier_difference
orders.isna().sum()
```

```python
customers_orders['customers_carrier_difference'] = customers_carrier_difference
```

```python
avg_deliveryy_by_city = customers_orders.groupby('customer_city')['customers_carrier_difference'].transform('mean')
avg_deliveryy_by_city
```

```python
delivered_customers_nulls = (orders['order_status'] == 'delivered') & (orders['order_delivered_customer_date'] == pd.to_datetime('1999-01-01 00:00:00'))&(orders['order_delivered_carrier_date'] != pd.to_datetime('1999-01-01 00:00:00'))
orders.loc[delivered_customers_nulls,'order_delivered_customer_date'] = orders.loc[delivered_customers_nulls,'order_delivered_carrier_date'] + avg_deliveryy_by_city
```
