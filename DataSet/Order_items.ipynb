```python
order_items['average_prices'] = order_items.groupby('product_id')['price'].transform('mean').round(2)
order_items
```

```python
price_Q1 = order_items['price'].quantile(0.25)
price_Q3 = order_items['price'].quantile(0.75)
```

```python
price_IQR = price_Q3 - price_Q1
price_lower_bound = price_Q1 - 1.5 * price_IQR
price_Upper_bound = price_Q3 + 1.5 * price_IQR
```

```python
price_lower_bound
```

```python
price_Upper_bound
```

```python
price_upper_limit = order_items['price'].quantile(0.95)
price_upper_limit
```

```python
price_lower_limit = order_items['price'].quantile(0.5)
price_lower_limit
```

```python
order_items.shape
```

```python
price_outliers = order_items[(order_items['price'] < price_lower_bound) |(order_items['price'] > price_Upper_bound)]
price_outliers
```

```python
price_outliers[price_outliers['product_id'] == 'fe59a1e006df3ac42bf0ceb876d70969']
```

```python
price_no_outliers = order_items[(order_items['price'] >= price_lower_bound) & (order_items['price'] <= price_Upper_bound)]
price_no_outliers
```

```python
# frights outliers
frights_Q1 = order_items['freight_value'].quantile(0.25)
frights_Q3 = order_items['freight_value'].quantile(0.75)
frights_IQR = frights_Q3 - frights_Q1
```

```python
frights_upper_bound = frights_Q1 - 1.5* frights_IQR
frights_lower_bound = frights_Q3 + 1.5*frights_IQR
```

```python
frights_no_outliers = order_items[(order_items['freight_value'] >= frights_lower_bound) & (order_items['freight_value'] <= frights_upper_bound)]
frights_no_outliers
```
